# Лабораторная работа №1. Ансамбли моделей

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import time
import matplotlib.pyplot as plt
import seaborn as sns
from bagging_classifier import ManualBaggingClassifier
```

## Загрузка и анализ данных

```python
df = pd.read_csv('breast-cancer.csv')

print("Первые 5 строк датасета:")
display(df.head())

print("\nИнформация о датасете:")
display(df.info())

print("\nСтатистическое описание числовых признаков:")
display(df.describe())
```

## Предобработка данных

```python
print("Количество пропущенных значений в каждом столбце:")
display(df.isnull().sum())

plt.figure(figsize=(8, 6))
df['diagnosis'].value_counts().plot(kind='bar')
plt.title('Распределение классов')
plt.xlabel('Класс')
plt.ylabel('Количество образцов')
plt.show()
```

```python
le = LabelEncoder()
y = le.fit_transform(df['diagnosis'])

X = df.drop(['id', 'diagnosis'], axis=1)

print("Размерность матрицы признаков:", X.shape)
print("Количество классов:", len(np.unique(y)))
print("\nСоответствие классов:")
for i, label in enumerate(le.classes_):
    print(f"{label}: {i}")
```

## Оценка производительности моделей

```python
n_estimators = 20
max_samples = 0.8

manual_bagging = ManualBaggingClassifier(n_estimators=n_estimators, max_samples=max_samples)

tree = DecisionTreeClassifier()
sklearn_bagging = BaggingClassifier(n_estimators=n_estimators, max_samples=max_samples, random_state=42)
sklearn_bagging.set_params(estimator=tree)
```

```python
start_time = time.time()
manual_scores = cross_val_score(manual_bagging, X, y, cv=5)
manual_time = time.time() - start_time

print("Ручная реализация:")
print(f"Средняя точность: {manual_scores.mean():.3f} (+/- {manual_scores.std() * 2:.3f})")
print(f"Время обучения: {manual_time:.2f} секунд")
```

```python
start_time = time.time()
sklearn_scores = cross_val_score(sklearn_bagging, X, y, cv=5)
sklearn_time = time.time() - start_time

print("\nРеализация sklearn:")
print(f"Средняя точность: {sklearn_scores.mean():.3f} (+/- {sklearn_scores.std() * 2:.3f})")
print(f"Время обучения: {sklearn_time:.2f} секунд")
```

Разовый запуск (проверочный)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
manual_bagging.fit(X_train, y_train)
y_pred = manual_bagging.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

## Сравнение результатов

```python
# Создание таблицы с результатами
results = pd.DataFrame({
    'Реализация': ['Ручная', 'Sklearn'],
    'Средняя точность': [manual_scores.mean(), sklearn_scores.mean()],
    'Стандартное отклонение': [manual_scores.std(), sklearn_scores.std()],
    'Время обучения (сек)': [manual_time, sklearn_time]
})

display(results)

# Визуализация результатов
plt.figure(figsize=(10, 6))
x = np.arange(2)
width = 0.35

plt.bar(x - width/2, results['Средняя точность'], width, label='Точность')
plt.bar(x + width/2, results['Время обучения (сек)'], width, label='Время обучения')

plt.xlabel('Реализация')
plt.xticks(x, ['Ручная', 'Sklearn'])
plt.title('Сравнение реализаций баггинга')
plt.legend()
plt.show()
```
